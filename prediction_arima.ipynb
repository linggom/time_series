```python
sales_data = sales['Weekly_Sales']
```

```python
sales_test = sales_data[-10:]
```

```python
sales_train = sales_data[:-10]
```

```python
predictions_arima = []
for i in range(sales_test.shape[0]):
    model = ARIMA(sales_train, order=(0,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    sales_train = sales_train.append(pd.Series(yhat, index=[sales_test.index[i]]))
    predictions_arima.append(yhat[0])
    print("label %s prediction %s "% (sales_test[i], yhat[0]))
predictions_arima = np.array(predictions_arima)
```

```python
math.sqrt(mean_squared_error(sales_test.values, predictions_arima))
```

```python
predictions_arma = []
for i in range(sales_test.shape[0]):
    model = ARMA(sales_train, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    sales_train = sales_train.append(pd.Series(yhat, index=[sales_test.index[i]]))
    predictions_arma.append(yhat[0])
    print("label %s prediction %s "% (sales_test[i], yhat[0]))
predictions_arma = np.array(predictions_arma)
```

```python
math.sqrt(mean_squared_error(sales_test.values, predictions_arma))
```
