```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sample_count = 4
```

# ARIMA 

```python
sales_test = sales_data[-sample_count:]
sales_train = sales_data[:-sample_count]
predictions_arima = []
for i in range(sales_test.shape[0]):
    model = ARIMA(sales_train, order=(0,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    sales_train = sales_train.append(pd.Series(yhat, index=[sales_test.index[i]]))
    predictions_arima.append(yhat[0])
    print(sales_test[i], yhat[0])
predictions_arima = np.array(predictions_arima)
math.sqrt(mean_squared_error(sales_test.values, predictions_arima))
```

# ARMA

```python
sales_test = sales_data[-sample_count:]
sales_train = sales_data[:-sample_count]
predictions_arma = []
for i in range(sales_test.shape[0]):
    model = ARMA(sales_train, order=(0,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    sales_train = sales_train.append(pd.Series(yhat, index=[sales_test.index[i]]))
    predictions_arma.append(yhat[0])
    print("label %s prediction %s "% (sales_test[i], yhat[0]))
predictions_arma = np.array(predictions_arma)
math.sqrt(mean_squared_error(sales_test.values, predictions_arma))
```

# AUTO ARIMA

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
from pyramid.arima import auto_arima
sales_test = sales_data[-sample_count:]
sales_train = sales_data[:-sample_count]

stepwise_model = auto_arima(sales_train, start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(stepwise_model.aic())
future_forecast = stepwise_model.predict(n_periods=sample_count)
math.sqrt(mean_squared_error(sales_test.values, future_forecast))
```

```python
pyplot.plot(sales_test.values)
pyplot.plot(predictions_arima, color='red')
pyplot.plot(predictions_arma, color='green')
pyplot.plot(future_forecast, color='yellow')

pyplot.show()
```
